```python
import yfinance as yf

# Define the ticker symbol
ticker = 'AAPL'  # Example: Apple Inc.

# Fetch the option chain data
option_chain = yf.Ticker(ticker).option_chain

# Get the data for a specific expiration date (e.g., 'YYYY-MM-DD')
expiration_date = '2024-07-12'
opt_chain = option_chain(expiration_date)

# Access calls and puts data
calls = opt_chain.calls  # Option calls data
puts = opt_chain.puts    # Option puts data

# Print out the option chain data
print("Calls:")
print(calls)
print("\nPuts:")
print(puts)
```

```python
import yfinance as yf
import pandas as pd


ticker = 'RS'  # Example: Reliance Steel & Aluminum Co.


option_chain = yf.Ticker(ticker).option_chain

expiration_date = '2024-07-19'
opt_chain = option_chain(expiration_date)


calls_df = pd.DataFrame(opt_chain.calls)
puts_df = pd.DataFrame(opt_chain.puts)

option_chain_df = pd.concat([calls_df, puts_df], keys=['Calls', 'Puts'])


csv_file = f"{ticker}_option_chain_{expiration_date}.csv"

option_chain_df.to_csv(csv_file, index=False)

print(f"Option chain data saved to {csv_file}")
```
